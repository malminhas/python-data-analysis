## Part A: The Jupyter Environment

### What is a Jupyter notebook?

A Jupyter notebook is a document that mixes three things in one place:

- **Code** that you can run interactively
- **Output** (numbers, tables, charts) that appears directly below each code cell
- **Explanatory text** written in Markdown (like this cell)

This is different from a `.py` script, which you run all at once from the terminal.
In a notebook you run one cell at a time and can inspect results before moving on.

### Cell types

There are two cell types you will use:

| Type | What it does | How to spot it |
|------|-------------|----------------|
| **Code** | Runs Python when you press Shift+Enter | Grey background, `[ ]:` on the left |
| **Markdown** | Renders formatted text | No bracket on the left when rendered |

### Essential keyboard shortcuts

| Shortcut | What it does |
|----------|-------------|
| **Shift + Enter** | Run the current cell and move to the next |
| **Ctrl + Enter** | Run the current cell and stay on it |
| **Esc** | Enter command mode (blue border) |
| **Esc then A** | Insert a new cell Above |
| **Esc then B** | Insert a new cell Below |
| **Esc then M** | Change cell to Markdown |
| **Esc then Y** | Change cell to Code |
| **Esc then D D** | Delete the current cell |

### B.1 Variables and data types

In Stata you might write:

```stata
gen age = 25
gen name = "Alice"
```

In Python the equivalent is simply:

````python
# Assign values to variables
age = 25            # integer (whole number)
height = 1.73       # float (decimal number)
name = "Alice"      # string (text, use single or double quotes)
is_student = True   # boolean (True or False)

# Print them all at once
print(age, height, name, is_student)
````

````python
# type() tells you what kind of value a variable holds
print(type(age))
print(type(height))
print(type(name))
print(type(is_student))
````

> **Stata comparison:** Stata distinguishes numeric and string variables at the dataset level.
> In Python, every individual value has a type, and you can mix types freely in your code.

### B.2 Lists and dictionaries

These are Python's two most common ways of grouping values together.

**A list** is an ordered sequence — like a Stata `levelsof` result or a column of values.
**A dictionary** maps keys to values — useful for lookup tables and structured records.

````python
# List: ordered, indexed from 0
countries = ["China", "United States", "Germany", "United Kingdom", "France"]

print("First country:  ", countries[0])   # index 0 = first item
print("Last country:   ", countries[-1])  # -1 = last item
print("Number of items:", len(countries))
````

````python
# Dictionary: key-value pairs
ev_sales_2025 = {
    "China":          13_300_000,
    "United States":   1_530_000,
    "Germany":           860_000,
    "United Kingdom":    700_000,
    "France":            450_000,
}

# Access a value by its key
print("China EV sales:", ev_sales_2025["China"])

# Iterate over the dictionary
for country, sales in ev_sales_2025.items():
    print(f"  {country}: {sales:,}")
````

> **Note on f-strings:** `f"  {country}: {sales:,}"` is a convenient way to embed variable values
> inside a string. The `:,` format code adds thousands separators.

### B.3 For loops

A loop runs the same block of code once for each item in a sequence.

````python
# Loop over a list
for country in countries:
    print(country)
````

````python
# Loop with a condition inside
for country in countries:
    if country.startswith("G"):
        print(country, "starts with G")
````

> **Stata comparison:** a Stata `foreach` loop works the same way:
> ```stata
> foreach country in "China" "Germany" {
>     display "`country'"
> }
> ```
> Python's syntax is just a little less verbose.

### B.4 Functions

A function is a reusable block of code.
You define it once with `def`, then call it as many times as you like.

````python
# Define a function
def ev_share(electric, non_electric):
    """Calculate electric cars as a percentage of total sales."""
    total = electric + non_electric
    return round(electric / total * 100, 1)


# Call the function
china_share = ev_share(13_300_000, 11_700_000)
print(f"China EV share: {china_share}%")

uk_share = ev_share(700_000, 1_300_000)
print(f"UK EV share:    {uk_share}%")
````

> The triple-quoted text inside the function (three quotes either side) is called a **docstring**.
> It documents what the function does. Run `help(ev_share)` to see it displayed.

## Part C: First Steps with pandas

pandas is the Python equivalent of Stata's data management environment.
Its central object is the **DataFrame** — a table of rows and columns.

| Stata concept | pandas equivalent |
|---------------|-------------------|
| Dataset (`.dta` file) | `DataFrame` |
| Variable | Column (`Series`) |
| Observation | Row |
| `use` | `pd.read_csv()` |
| `describe` | `df.info()` |
| `summarize` | `df.describe()` |
| `list in 1/5` | `df.head()` |
| `tabulate var` | `df["var"].value_counts()` |

### C.1 Loading data

```stata
* Stata
use "car-sales.dta", clear
```

````python
# Python — load a CSV file into a DataFrame
df = pd.read_csv("car-sales.csv")

# Quick check: did it load?
print(f"Loaded {len(df)} rows and {len(df.columns)} columns.")
````

> **What is `df`?** Just a variable name — short for DataFrame.
> You could call it anything (`cars`, `data`, `my_table`), but `df` is the Python convention.

### C.2 Inspecting the structure

```stata
* Stata
describe
```

````python
# Shape: (rows, columns)
print("Shape:", df.shape)
````

````python
# Column names and data types — like Stata's 'describe'
df.info()
````

````python
# See the first 5 rows — like Stata's 'list in 1/5'
df.head()
````

````python
# See the last 5 rows
df.tail()
````

### C.3 Summary statistics

```stata
* Stata
summarize
```

````python
# Summary statistics for all numeric columns — like Stata's 'summarize'
df.describe()
````

````python
# Summary statistics for a single column
df["Electric cars"].describe()
````

> **Column access syntax:** `df["Electric cars"]` selects one column by name.
> The result is a pandas **Series** — a single column with an index.
> In Stata terms, it is like referring to a single variable.

### C.4 Frequency counts

```stata
* Stata
tabulate Entity
```

````python
# Count observations per country/entity — like Stata's 'tabulate'
df["Entity"].value_counts()
````

````python
# How many unique entities are there?
print("Unique entities:", df["Entity"].nunique())

# What are they?
print(sorted(df["Entity"].unique()))
````

### C.5 Filtering rows

```stata
* Stata
list if Entity == "World"
```

````python
# Select only the rows where Entity is "World"
world = df[df["Entity"] == "World"]
world
````

````python
# Filter on two conditions at once with & (and)
# Keep only World rows from 2020 onwards
df[(df["Entity"] == "World") & (df["Year"] >= 2020)]
````

### C.6 Selecting columns

```stata
* Stata
keep Entity Year electric_cars
```

````python
# Select a single column — returns a Series
df["Year"]
````

````python
# Select multiple columns — pass a list, returns a DataFrame
df[["Entity", "Year", "Electric cars"]].head(10)
````

### C.7 Checking for missing values

```stata
* Stata
misstable summarize
```

````python
# Count missing values in each column
df.isnull().sum()
````

````python
# Which rows have a missing 'Non-electric cars' value?
df[df["Non-electric cars"].isnull()]
````

## Quick Reference: Stata to pandas

| Task | Stata | Python / pandas |
|------|-------|-----------------|
| Load CSV | `import delimited file.csv` | `df = pd.read_csv("file.csv")` |
| Rows and columns | `describe` | `df.shape` |
| Variable types | `describe` | `df.info()` |
| Summary stats | `summarize` | `df.describe()` |
| First N rows | `list in 1/5` | `df.head(5)` |
| Frequency table | `tabulate var` | `df["var"].value_counts()` |
| Unique values | `levelsof var` | `df["var"].unique()` |
| Count unique | `tabulate var` | `df["var"].nunique()` |
| Filter rows | `keep if age > 18` | `df[df["age"] > 18]` |
| Select columns | `keep var1 var2` | `df[["var1", "var2"]]` |
| Missing count | `misstable summarize` | `df.isnull().sum()` |
